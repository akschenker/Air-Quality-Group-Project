```python
# Dependencies
import requests
import pymongo
import pandas as pd
import numpy as np


from pprint import pprint

import config

cities_pd = pd.read_csv("500cities.csv")
```

```python
cities = cities_pd[['City', 'Population', 'Latitude', 'Longitude']]
cities.head()
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.air_quality
collection = db.air_results
```

```python
row_count = 0
api_key = config.API_key

    
params = {"lat": lat, "lon": lon,"key": api_key}


url = "https://api.breezometer.com/air-quality/v2/current-conditions?"


for index, row in cities.iterrows():
    try:
        lat = row['Latitude']
        lon = row['Longitude']
    
        params['lat'] = lat
        params['lon'] = lon
    
        query_url = f"{url}lat={lat}&lon={lon}&key={api_key}&features=breezometer_aqi,local_aqi,health_recommendations,sources_and_effects,pollutants_concentrations,pollutants_aqi_information"
        response = requests.get(query_url).json()
        
        aqi = response['data']['indexes']['baqi']['aqi']
        category = response['data']['indexes']['baqi']['category']
        health_recom_general = response['data']['health_recommendations']['general_population']
        health_recom_active = response['data']['health_recommendations']['active']
        pollutant_obj = dict()
        for pollutant in response['data']['pollutants'].items():
            pollutant_display_name = pollutant[1]['display_name']
            pollutant_full_name = pollutant[1]['full_name']
            pollutant_concentration = pollutant[1]['concentration']['value']
            pollutant_units = pollutant[1]['concentration']['units']
            pollutant_name = pollutant[0]
            pollutant_obj[pollutant_name] = {
                    'pollutant_display_name': pollutant_display_name,
                    'pollutant_full_name': pollutant_full_name,
                    'pollutant_concentration': pollutant_concentration,
                    'pollutant_units': pollutant_units}
   
    
        post = {
                'city': row['City'],
                'population': row['Population'],
                'latitude': row['Latitude'],
                'longitude': row['Longitude'],
                'aqi': aqi,
                'air_category': category,
                'health_recom_general': health_recom_general,
                'health_recom_active': health_recom_active,
                'pollutant': pollutant_obj

             }

        collection.insert_one(post)
    
    except Exception as e:
        print(e)



#     pprint(response)
#     print(query_url)
```
